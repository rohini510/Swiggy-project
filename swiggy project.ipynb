4.Which are the top 10 restaurants with fast delivery in Bangalore city

```python
df[df['City'] == 'Bangalore'].sort_values(['Restaurant', 'Delivery time'],ascending=[False,False]).head(10)
```

```python


# Filter the dataset for restaurants in Bangalore
bangalore_restaurants = df[df['City'] == 'Bangalore']

# Sort the filtered data by delivery time in ascending order
fastest_bangalore_restaurants = bangalore_restaurants[['Restaurant', 'Delivery time']].sort_values(by='Delivery time', ascending=True)

# Get the top 10 restaurants with the fastest delivery times
top_10_fastest_bangalore = fastest_bangalore_restaurants.head(10)

print(top_10_fastest_bangalore)
```
